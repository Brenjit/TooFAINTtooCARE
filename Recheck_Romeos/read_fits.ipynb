```python
from astropy.io import fits

# === Define the file path ===
file_path = '/Volumes/MY_SSD_1TB/Work_PhD/July-August/CEERS_data/nircam1/egs_all_acs_wfc_f606w_030mas_v1.9_nircam1_mef.fits'  # üîÅ Replace with actual path

# === Print general info about all HDUs ===
print("üîç FITS File Structure:")
fits.info(file_path)
print("=" * 60)

# === Open the file and inspect each HDU ===
with fits.open(file_path) as hdul:
    for idx, hdu in enumerate(hdul):
        print(f"\n--- HDU {idx} ---")
        print(f"Type: {type(hdu)}")
        print(f"Header:\n{hdu.header}")
        
        # Try to print data shape/type if present
        if hdu.data is not None:
            print(f"Data type: {type(hdu.data)}")
            print(f"Data shape: {hdu.data.shape}")
        else:
            print("No data in this HDU.")

```
